```python
import ccxt
from crypto_data_fetcher.binance_future import BinanceFutureFetcher
import joblib
import pandas as pd
```

## crypto_data_fetcher
https://github.com/richmanbtc/crypto_data_fetcher/blob/master/crypto_data_fetcher/binance_future.py

binanceはccxtの関数を利用

- [ccxt](https://github.com/ccxt/ccxt)
- [ccxt python](https://github.com/ccxt/ccxt/tree/master/doc#python-1)

```python
memory = joblib.Memory('/tmp/binance_future_fetcher_cache', verbose=0)
fetcher = BinanceFutureFetcher(ccxt_client=ccxt.binance())

df = fetcher.fetch_ohlcv(
    market='BTCUSDT', # 市場のシンボルを指定
    interval_sec=15 * 60, # 足の間隔を秒単位で指定。e.g.15分足
)

# 実験に使うデータ期間を限定
df = df[df.index < pd.to_datetime('2021-04-01 00:00:00Z')]

display(df)
df.to_pickle('df_ohlcv.pkl')
```
